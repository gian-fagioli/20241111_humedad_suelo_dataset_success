```python
import pandas as pd
import pickle
```

```python
with open('planet_create_subscription_result_It1.pkl', 'rb') as archivo:
    subscription_result_It1 = pickle.load(archivo)

with open('planet_create_subscription_result_It2.pkl', 'rb') as archivo:
    subscription_result_It2 = pickle.load(archivo)
```

```python
data = [{'plot_id': sub['name'].split('_')[0], 'sub_id': sub['id'], 'name': sub['name']} for sub in subscription_result_It1]

# Crear el DataFrame a partir de los datos extra√≠dos
df = pd.DataFrame(data)

data_2 = [{'plot_id': sub['name'].split('_')[0], 'sub_id': sub['id'], 'name': sub['name']} for sub in subscription_result_It2]
df_2 = pd.DataFrame(data_2)
```

```python
df = pd.concat([df, df_2], ignore_index=True)
```

```python
df.to_csv('subs_plotids.csv', index=False)
```

```python
path_prefix = 'Soil_Water_Holding_Capacity/SWHC/'
aws_url = f's3://kilimo-planet-labs-soil-water-content/{path_prefix}'

with open('aws_commands_download_images.txt', 'a') as the_file:
    
    for index, row in df.iterrows():
        print(row['sub_id'])
        sub_id = row['sub_id']
        aws_command = f'aws s3 sync {aws_url}{sub_id} ./SWC_images/{path_prefix}{sub_id}/'
        the_file.write(f'{aws_command}\n')
```
