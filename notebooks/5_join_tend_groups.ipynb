```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_tend = pd.read_csv('magnitud_pendiente.csv')
```

```python
df_tend.head(3)
```

```python
df_dataset = pd.read_csv('dataset_agro_cultiv_reg.csv')
```

```python
df_dataset.dtypes
```

```python
# Join dfs
df = df_tend.merge(df_dataset, on='plot_id')
```

```python
df.head()
```

```python
sns.set_style("darkgrid")

sns.histplot(data=df, x="Pendiente_tendencia", hue="Tipo de manejo del campo/lote",
             element="poly", palette=sns.color_palette("tab10"))
```

```python
sns.violinplot(data=df,y="Pendiente_tendencia", x="Tipo de manejo del campo/lote")
```
